```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

        Load data     

```python
train = pd.read_csv('train.csv')
```

        Basic settings        

```python
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

        Basic age distribution     

```python
train['Age'].hist(bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.savefig('age_distribution.png')
plt.show()
```

        Survival histogram     

```python
train['Survived'].hist(bins=2, rwidth=0.8)
plt.title('Survival Distribution')
plt.xlabel('Survived (0=No, 1=Yes)')
plt.xticks([0.25, 0.75], ['0', '1'])
plt.ylabel('Count')
plt.savefig('survival_distribution.png')
plt.show()
```

        Scatter plot: Age vs Fare      

```python
plt.scatter(train['Age'], train['Fare'], alpha=0.5, s=10)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.savefig('age_vs_fare.png')
plt.show()
```

        Box plot        

```python
train[['Age', 'Fare']].plot.box()
plt.title('Box Plot: Age and Fare')
plt.ylabel('Value')
plt.savefig('box_plot.png')
plt.show()
```

        KDE plot for age        

```python
train['Age'].plot.kde()
plt.title('Age Density Plot')
plt.xlabel('Age')
plt.ylabel('Density')
plt.savefig('age_density.png')
plt.show()
```

        Hexbin plot     

```python
train.plot.hexbin('Age', 'Fare', gridsize=20)
plt.title('Age vs Fare Density')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.savefig('hexbin_plot.png')
plt.show()
```

        Scatter matrix (simplified)     

```python
pd.plotting.scatter_matrix(train[['Age', 'Fare', 'Pclass']], figsize=(10, 8))
plt.suptitle('Scatter Matrix', y=1)
plt.savefig('scatter_matrix.png')
plt.show()
```

        Multiple subplots       
                Subplot 1: Age histogram                
                Subplot 2: Fare histogram               
                Subplot 3: Survival bar plot            
                Subplot 4: Scatter Age vs Pclass        

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

train['Age'].hist(bins=20, ax=axes[0, 0])
axes[0, 0].set_title('Age Distribution')

train['Fare'].hist(bins=20, ax=axes[0, 1])
axes[0, 1].set_title('Fare Distribution')

train['Survived'].value_counts().plot(kind='bar', ax=axes[1, 0])
axes[1, 0].set_title('Survival Count')
axes[1, 0].set_xlabel('Survived')
axes[1, 0].set_xticklabels(['No', 'Yes'], rotation=0)

axes[1, 1].scatter(train['Age'], train['Pclass'], alpha=0.5, s=10)
axes[1, 1].set_title('Age vs Pclass')
axes[1, 1].set_xlabel('Age')
axes[1, 1].set_ylabel('Pclass')

plt.tight_layout()
plt.savefig('multi_plot.png')
plt.show()
```

        Display basic statistics        

```python
print("Dataset shape:", train.shape)
print("\nColumn names:", list(train.columns))
print("\nBasic statistics:")
print(train[['Age', 'Fare', 'Pclass']].describe())
```

        Save processed data     

```python
train.to_csv('processed_data.csv', index=False)
print("\nData saved to 'processed_data.csv'")
print("Visualization complete.")
```
