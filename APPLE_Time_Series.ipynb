```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('AAPL.csv')
data.drop('index', inplace= True, axis= 1)
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
pd.to_datetime("2021-11-23")
```

```python
data['Date']= data['Date'].apply(pd.to_datetime)
data.info()
```

```python
data.set_index('Date', inplace= True)
```

```python
data['Open'].plot()
```

```python
data['Volume'].plot()
```

```python
data.resample(rule= 'M').mean()['Open'].plot()
```

```python
data.resample(rule= 'Y').mean()['Volume'].plot(kind= 'bar')
```

```python
plt.scatter(data['Adj Close'], data['Volume'])
```

```python
from bokeh.plotting import output_file, figure, show

p= figure()
p.line(data['Open'], data['Volume'], color= 'blue', legend_label= 'line')
p.legend.click_policy= 'hide'
show(p)
```
